Question: Imagine a scenario where an investor wants to understand the risks and returns associated with various stocks before investing in them.  Given two stocks, find out which stock is at lesser risk by looking into measures of dispersion of stocks values. Also implement the 12 sub task given in todayâ€™s class for the same problem

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline

beml_df=pd.read_csv('BEML.csv')
glaxo_df=pd.read_csv('GLAXO.csv')

print(beml_df.head(5))
print(glaxo_df.head(5))

beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
print(beml_df)
print(glaxo_df)

beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))

beml_df['Gain']=beml_df['Close'].pct_change(periods=1)
glaxo_df['Gain']=glaxo_df['Close'].pct_change(periods=1)
print(beml_df.Gain.head(5))
print(glaxo_df.Gain.head(5))

plt.plot(beml_df['Close'],label='BEML')
plt.plot(glaxo_df['Close'],label='GLAXO',color='r')
plt.xlabel('Time')
plt.ylabel('Close')
plt.legend()
plt.show()

sn.distplot(beml_df.Gain,hist=False)
sn.distplot(glaxo_df.Gain,hist=False,color='r')


beml_gain=beml_df.Gain.mean()
beml_SD=beml_df.Gain.std()
beml_var=beml_df.Gain.var()
print("Bemlt mean: ",beml_gain)
print("Beml StD: ",beml_SD)
print("Beml VAr: ",beml_var)

glaxo_gain=glaxo_df.Gain.mean()
glaxo_SD=glaxo_df.Gain.std()
glaxo_var=glaxo_df.Gain.var()
print("Glaxo mean: ",glaxo_gain)
print("Glaxo StD: ",glaxo_SD)
print("Glaxo VAr: ",glaxo_var)

print('Conclusion: Glaxo Stock is better to buy as the data is not scattred much and it ins not varing much.')
```
